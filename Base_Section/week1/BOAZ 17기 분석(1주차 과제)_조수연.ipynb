### 6. grade별로 평균 가격을 보여주세요.

```python
grade_mean = data.groupby('grade')['price'].agg([('grade_mean', 'mean')])
grade_mean
```

### 7. 6번에서 만든 데이터를 원본 데이터와 merge해주세요.

```python
pd.merge(data, grade_mean, on = 'grade', how = 'inner')
```

### 9. 이후 Price를 예측하는 모델을 제작하기 위해 추가적인 변수를 만드려고 합니다. Price의 데이터 분포를 확인하고 1개 이상의 EDA를 통해 의견을 제시해주세요.

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.rc('font', family='malgun gothic')
plt.rc('axes', unicode_minus=False)
```

**price**

```python
data.price.describe()
```

```python
plt.boxplot(data.price)
```

비싼 값의 집들이 많이 분포하고 있음

**(1) 침실 개수에 따른 price**

```python
plt.scatter(data.bedrooms, data.price)
plt.xlabel('bedrooms')
plt.ylabel('price')
```

침실 개수가 5~6개일 때 price가 가장 높다.

**(2) 화장실 개수에 따른 price**

```python
plt.scatter(data.bathrooms, data.price)
plt.xlabel('bathrooms')
plt.ylabel('price')
```

화장실 개수가 늘어남에 따라 price가 오르는 경향이 있다.

**(3) sqft_living에 따른 price**

```python
plt.scatter(data.sqft_living, data.price)
plt.xlabel('sqft_living')
plt.ylabel('price')
```

sqft_living가 높을수록 price가 높다.

**(4) 침실+화장실 개수에 따른 price**

```python
data['rooms'] = data.bedrooms + data.bathrooms
data.rooms
plt.scatter(data.rooms, data.price)
plt.xlabel('rooms')
plt.ylabel('price')
```
